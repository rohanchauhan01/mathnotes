## Phase Shift

y = $Amplitude\cdot sin(\frac{2\pi}{period}x + horizontal\,shift) + vertical\,shift$

horizontal shift: negative for right, positive for left


```python
#import pylab
%pylab inline
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi)
y = np.sin(x)
pi = np.pi
plt.title('sine wave $sin(x)$')
plt.xticks( [-2*pi, -1.5*pi, -pi, -pi/2, 0,pi/2,pi,1.5*pi,2*pi],
        [r'$-2\pi$', r'$-\frac{3\pi}{2}$',r'$-\pi$',r'$-\frac{\pi}{2}$' ,'$0$',r'$+\frac{pi}{2}$',r'$+\pi$', r'$\frac{3\pi}{2}$', r'$+2\pi$'])
plt.plot(x,y)

#$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$']
```

```python
x = np.linspace(-1 * np.pi, 1 * np.pi)
y = sin(2*x*np.pi/(1*pi))
pi = np.pi
plt.title('sine wave $sin(x)+\frac{pi}{2}$')
#plt.xticks( [-2*pi, -1.5*pi, -pi, -pi/2, 0,pi/2,pi,1.5*pi,2*pi],
#        [r'$-2\pi$', r'$-\frac{3\pi}{2}$',r'$-\pi$',r'$-\frac{\pi}{2}$' ,'$0$',r'$+\frac{pi}{2}$',r'$+\pi$', r'$\frac{3\pi}{2}$', r'$+2\pi$'])
plt.plot(x,y)

#$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$']
```
